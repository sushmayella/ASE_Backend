## Train Data Preparation

```python
categories=os.listdir(TRAIN_PATH)
print('Categories Name :' ,categories)
labels=[i for i in range(len(categories))]
print('After convert it to 0,1,2,3 :',labels)
```

```python
label_dict=dict(zip(categories,labels))
print(label_dict)
print('len of Labels dictionary : ' , len(label_dict))
```

## Collecting Training images and Labels

```python
Imges_Data_Train=[]
target_Train=[]


for  i in categories :
    path_Cat=os.path.join(TRAIN_PATH,i)
    print('Path Now is :', path_Cat)
    Img_Names=os.listdir(path_Cat)
    
    for img in Img_Names :
        img_path=os.path.join(path_Cat,img)
        imge=cv2.imread(img_path)
        gray_img=cv2.cvtColor(imge,cv2.COLOR_BGR2GRAY)           
        dim=(IMG_SIZE,IMG_SIZE)
        resized_img = cv2.resize(gray_img, dim)
        Imges_Data_Train.append(resized_img)
        target_Train.append(label_dict[i])
```

```python
print(target_Train)
```

## Data Normalization and Reshaping

```python
Imges_Data_Train=np.array(Imges_Data_Train)/255.0
Imges_Data_Train=np.reshape(Imges_Data_Train,(Imges_Data_Train.shape[0],IMG_SIZE,IMG_SIZE,1))
```

```python
target_Train=np.array(target_Train)
```

## Class Size Setting

```python
Target_Train=to_categorical(target_Train)
print(Target_Train)
```

## Test Data preparation

```python
categories=os.listdir(TEST_PATH)
print('Categories Name :' ,categories)
labels=[i for i in range(len(categories))]
print('After convert it to 0,1,2,3 :',labels)
```

```python
label_dict=dict(zip(categories,labels))
print(label_dict)
print('len of Labels dictionary : ' , len(label_dict))
```

```python
Imges_Data_Test=[]
target_Test=[]

for  i in categories :
    path_Cat=os.path.join(TEST_PATH,i)
    print('Path Now is :', path_Cat)
    Img_Names=os.listdir(path_Cat)
    
    for img in Img_Names :
        img_path=os.path.join(path_Cat,img)
        imge=cv2.imread(img_path)
        gray_img=cv2.cvtColor(imge,cv2.COLOR_BGR2GRAY)           
        dim=(IMG_SIZE,IMG_SIZE)
        resized_img = cv2.resize(gray_img, dim)
        Imges_Data_Test.append(resized_img)
        target_Test.append(label_dict[i])
```

```python
print(target_Test)
```

## Visualization

### pie-chart

```python
plt.figure(figsize=(8,8))
plt.rcParams.update({"font.size":20})
pie_data=[len(target_Train),len(target_Test)]
labels=["Train_Data_Size","Test_Data_Size"]
colors=["teal","gold"]
plt.pie(x=pie_data,labels=labels,colors=colors,autopct="%.2f")
plt.title("Comparision of size of Train and Test Data")
plt.show()
```

## Data Normaization and Reshaping

```python
Imges_Data_Test=np.array(Imges_Data_Test)/255.0
Imges_Data_Test=np.reshape(Imges_Data_Test,(Imges_Data_Test.shape[0],IMG_SIZE,IMG_SIZE,1))
```

```python
target_Test=np.array(target_Test)
```

## Class Size Setting

```python
Target_Test=to_categorical(target_Test)
print(Target_Test)
```

## CNN(convolutional Neural Network

```python
model=Sequential()

model.add(Conv2D(64,(3,3),input_shape=Imges_Data_Train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(128,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Conv2D(128,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(4,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

## Model Building

```python
results=model.fit(Imges_Data_Train,Target_Train,epochs=20,batch_size=64,validation_data=(Imges_Data_Test,Target_Test))
```

## Accuracy and Loss Graphs

```python
plt.figure(figsize=(18,8))
plt.plot(results.history["accuracy"],color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.plot(results.history["val_accuracy"],color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.title("Accuract and Val_Accuracy Plot Graphs",fontsize=20)
plt.legend(["Accuracy","Val_Accuracy"])
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.plot(results.history["loss"],color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.plot(results.history["val_loss"],color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.title("Loss and Val_Loss Plot Graphs",fontsize=20)
plt.legend(["Accuracy","Loss"])
plt.show()
```

## Model Prediction

```python
CNN_prediction=model.predict(Imges_Data_Test,batch_size=64,verbose=1)
```

```python
print(CNN_prediction)
```

```python
y_predict = []
for i in range(0, len(CNN_prediction)):
    y_predict.append(int(np.argmax(CNN_prediction[i])))
len(y_predict)
```

```python
print(y_predict)
```

```python
y_true = []
for i in range(0, len(Target_Test)):
    y_true.append(int(np.argmax(Target_Test[i])))
len(y_true)
```

```python
print(y_true)
```

## Result Analysis

```python
from sklearn.metrics import classification_report,confusion_matrix
```

### Classification Report

```python
print(classification_report(y_true=y_true,y_pred=y_predict,target_names=categories))
```

### Confusion Matrix

```python
from mlxtend.plotting import plot_confusion_matrix
ax=plot_confusion_matrix(conf_mat=confusion_matrix(y_true=y_true,y_pred=y_predict),
                        figsize=(5,5),
                        class_names=categories,
                        cmap=plt.cm.Blues)
plt.title(label="Confusion Matrix")
plt.xticks(rotation=90)
plt.show()
```

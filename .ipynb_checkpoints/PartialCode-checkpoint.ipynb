## visualization

### Images from Train directory

```python
Apple___Apple_scab = 'input/Train/Apple___Apple_scab/'
Apple___Black_rot = 'input/Train/Apple___Black_rot/'
Apple___Cedar_apple_rust = 'input/Train/Apple___Cedar_apple_rust/'
Apple___healthy = 'input/Train/Apple___healthy/'
```

```python
dirlist=[Apple___Apple_scab, Apple___Black_rot,Apple___Cedar_apple_rust,Apple___healthy]
filepaths=[]
for i in dirlist:
    filelist=os.listdir(i)
    for f in filelist:
        filepath=os.path.join (i,f)
        filepaths.append(filepath)
```

```python
plt.figure(figsize=(12,8))
for i in range(25):
    random = np.random.randint(1,len(filepaths))
    plt.subplot(5,5,i+1)
    img=cv2.imread(filepaths[random])
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([]) 
plt.show()
```

## Visualizing train data's classes size

```python
Apple___Apple_scab = 'input/Train/Apple___Apple_scab/'
Apple___Black_rot = 'input/Train/Apple___Black_rot/'
Apple___Cedar_apple_rust = 'input/Train/Apple___Cedar_apple_rust/'
Apple___healthy = 'input/Train/Apple___healthy/'
```

```python
Apple___Apple_scab=os.listdir(Apple___Apple_scab)
Apple___Black_rot=os.listdir(Apple___Black_rot)
Apple___Cedar_apple_rust=os.listdir(Apple___Cedar_apple_rust)
Apple___healthy=os.listdir(Apple___healthy)
```

```python
plt.figure(figsize=(12,8))
plt.rcParams["font.size"]=20
labels=["Apple___Apple_scab","Apple___Black_rot","Apple___Cedar_apple_rust","Apple___healthy"]
train_data=[len(Apple___Apple_scab),len(Apple___Black_rot),len(Apple___Cedar_apple_rust),len(Apple___healthy)]
colors=["lime","pink","teal","green"]
plt.bar(x=labels,height=train_data,color=colors)
plt.xticks(rotation=90)
plt.show()
```

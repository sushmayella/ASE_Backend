# TESTING PROCESS

## Importing Libraries

```python
from tensorflow.keras.models import model_from_json
from pathlib import Path
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## Labels dictionary

```python
class_names = {0:'Apple___Apple_scab',
               1:'Apple___Black_rot', 
               2:'Apple___Cedar_apple_rust',
               3:'Apple___healthy'}
```

## Model Loading

```python
file_path = Path("model/cnn_model_structure.json")
model_structure = file_path.read_text()
model = model_from_json(model_structure)
```

```python
model.load_weights("model/cnn_model_weights.h5")
```

## User Input Image

```python
IMG_SIZE = 256
```

```python
img_path="UserInput/scab/scab1.jpg"
```

```python
imge=cv.imread(img_path)
gray_img=cv.cvtColor(imge,cv.COLOR_BGR2GRAY)           
dim=(IMG_SIZE,IMG_SIZE)
resized_img = cv.resize(gray_img, dim)
```

```python
plt.imshow(resized_img,cmap="gray")
plt.show()
```

## Normalization

```python
image_to_test = image.img_to_array(resized_img)/255.0
```

```python
list_of_images = np.expand_dims(image_to_test, axis=0)
```

```python
results = model.predict(list_of_images)
```

```python
print(results)
```

```python
single_result=results[0]
```

```python
print(single_result)
```

```python
most_likely_class_index = int(np.argmax(single_result))
print(most_likely_class_index)
```

```python
class_likelihood = single_result[most_likely_class_index] 
print(class_likelihood )
```

```python
class_label = class_names[most_likely_class_index]
print(class_label)
```

## Final Result

```python
print("Predicted class is  {} - Confidence Score: {:2f}".format(class_label, class_likelihood))
```
